librosa library is used to have constant sample rate for the audio as in general audio get saved at different sample rate but librosa present it in range of 21KHz  to 23KHz and also it does normalize the data in -1 to 1 

in mono channel audio there is only one signal that means data will be reprented in single array only but in sterio channel data will be presented in dual for many channels 

sample rate = the number of samples per second that are taken of a waveform to create a discete digital signal.
